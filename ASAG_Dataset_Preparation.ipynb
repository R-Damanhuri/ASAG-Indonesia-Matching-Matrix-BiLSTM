# MPI

```python
mpi = pd.read_excel("/content/data_mpi.xlsx")
```

```python
mpi = mpi.drop(columns=['NIM'])
```

```python
mpi = mpi.rename(columns={
    'Nilai Sae':'Nilai_1',
    'Nilai Iffa ':'Nilai_2'
})
```

```python
mpi['Rata-rata'] = (mpi['Nilai_1'] + mpi['Nilai_2'])/2
```

```python
mpi.insert(0,"Soal", "Dalam penelitian terdapat 3 aspek utama yang harus dimiliki dalam sebuah penelitian.  Sebutkan dan jelaskan selengkap mungkin yang anda ketahui mengenai hal tersebut!")
```

```python
mpi
```

```python
import re

def clean(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = ' '.join(text.split())

    return text
```

```python
mpi['Soal'] = mpi['Soal'].str.lower().apply(clean)
mpi['Jawaban'] = mpi['Jawaban'].str.lower().apply(clean)
```

```python
mpi.describe()
```
