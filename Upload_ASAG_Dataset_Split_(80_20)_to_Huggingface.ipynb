# ASAG Dataset

```python
import pandas as pd

asag = pd.read_csv("/content/ASAG-Indonesia-Matching-Matrix-BiLSTM/data/asag_cleaned.csv")
```

```python
asag
```

```python
from datasets import Dataset, DatasetDict

# Ubah DataFrame ke dalam format Dataset
asag_dataset = Dataset.from_pandas(asag)

# Hitung jumlah baris dalam dataset
num_rows = len(asag_dataset)

# Bagi dataset menjadi train dan validation
split_ratio = 0.8  # Anda dapat mengatur rasio sesuai kebutuhan
split_point = int(num_rows * split_ratio)

train_dataset = asag_dataset.select(indices=list(range(split_point)))
validation_dataset = asag_dataset.select(indices=list(range(split_point, num_rows)))

# Buat DatasetDict
asag_dict = DatasetDict({
    "train": train_dataset,
    "validation": validation_dataset
})
```

```python
asag_dict
```

```python
asag_dict.push_to_hub("damand2061/asag_cleaned", private=True)
```
